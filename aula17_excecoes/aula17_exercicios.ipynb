```python
#039.: Escreva um programa que realiza a soma das matrizes ou retorna um erro caso as dimensões delas não sejam compatíveis.

###############################################################################################################################
def leMat(a):
    m=int(input("Digite o número de linhas da Matriz: "))
    n=int(input("Digite o número de colunas da Matriz: "))
    
    mat=[[0 for i in range(n)] for j in range(m)]
    
    for i in range(m):
        for j in range(n):
            mat[i][j]=float(input("mat{}[{}][{}]=".format(a, i+1, j+1)))
    
    return mat
###############################################################################################################################
def printMat(mat):
    
    for i in range(len(mat)):
        for j in range(len(mat[0])):
            print(mat[i][j], end='   ')
        print()

###############################################################################################################################
def somaMat(m1, m2):
    if (len(m1)!=len(m2)) or (len(m1[0])!=len(m2[0])):
        raise ValueError("Dimensões incompatíveis com soma matricial")
        
    m3=[[0 for i in range(len(m1[0]))] for j in range(len(m1))]
    
    for i in range(len(m3)):
        for j in range(len(m3[0])):
            m3[i][j]=m1[i][j]+m2[i][j]
            
    return m3

###############################################################################################################################
print("Leitura das Matrizes")
mat1=leMat(1)
mat2=leMat(2)

try:
    printMat(somaMat(mat1, mat2))
    
except ValueError as msg:
    print(msg)
    
```
