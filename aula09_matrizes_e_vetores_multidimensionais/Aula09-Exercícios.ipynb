024.: Faça um programa para realizar operações com matrizes que tenha as seguintes funcionalidades: 

um menu para escolher a operação a ser realizada:

    leitura matriz 1
    leitura matriz 2
    impressao da matriz 1 e da matriz 2
    cálculo da soma da matriz 1 com a matriz 2 e impressao do resultado
    calculo da subtracao da matriz 1 com matriz 2 e impressao do resultado
    impressao da transposta da matriz 1 e da matriz 2

```python
##############################################################

def main():
    opcao=0
    while opcao!=9:
        print("1. Leitura da Matriz 1")
        print("2. Leitura da Matriz 2")
        print("3. Impressão das matrizes 1 e 2")
        print("4. Soma das matrizes 1 e 2")
        print("5. Multiplicação das matrizes 1 e 2")
        print("6. Subtração das matrizes 1 e 2")
        print("7. Transposta das matrizes 1 e 2")
        print("8. Sair")
        
        opcao=int(input("Digite o número da opção que deseja:"))
        
        if opcao==1:
            print()
            print("Matriz 1")
            mat1=leMatriz(1)
            
        elif opcao==2:
            print()
            print("Matriz 2")
            mat2=leMatriz(2)
            
        elif opcao==3:
            print()
            print("Matriz 1")
            printMatriz(mat1)
            print()
            print("Matriz 2")
            printMatriz(mat2)
            
        elif opcao==4:
            print()
            print("M1 + M2")
            somaMatriz(mat1, mat2)
                        
        elif opcao==5:
            print()
            print("M1 x M2")
            multMatriz(mat1, mat2)
            
        elif opcao==6:
            print()
            print("M1-M2")
            subtMatriz(mat1, mat2)
        
        elif opcao==7:
            print()
            print("M1t")
            transMatriz(mat1)
            print()
            print("M2t")
            transMatriz(mat2)
        
        elif opcao==8:
            print()
            print("Saindo")
            break
            
        print()
                        
        
############################################################## 

def leMatriz(a):
    n=int(input("Digite a dimensão da matriz quadrada: "))
    mat=[[]for i in range(n)]
    
    for i in range(n):
        for j in range(n):
            aux=float(input('Mat{}[{}][{}]='.format(a, i+1, j+1)))
            mat[i].append(aux)
    
    return mat

##############################################################

def printMatriz(mat):
    for i in range(len(mat)):
        for j in range(len(mat)):
            print(mat[i][j], end=' ')
        print()

############################################################## 

def somaMatriz(mat1, mat2):
    mat3=[[0 for j in range(len(mat1))]for i in range(len(mat1))]
    
    for i in range(len(mat1)):
        for j in range(len(mat1)):
            mat3[i][j]=mat1[i][j]+mat2[i][j]
    
    printMatriz(mat3)
    

############################################################## 

def subtMatriz(mat1, mat2):
    mat3=[[0 for j in range(len(mat1))]for i in range(len(mat1))]
    
    for i in range(len(mat1)):
        for j in range(len(mat1)):
            mat3[i][j]=mat1[i][j]-mat2[i][j]
    
    printMatriz(mat3)
    
##############################################################

def multMatriz(mat1, mat2):
    mat3=[[0 for j in range(len(mat1))]for i in range(len(mat1))]
    
    for i in range(len(mat1)):
        for j in range(len(mat1)):
            for k in range(len(mat1)): 
                mat3[i][j]=mat3[i][j] + (mat1[i][k]*mat2[k][j])
                
    printMatriz(mat3)
    
##############################################################

def transMatriz(mat):
    n=len(mat)
    matT=[[0 for j in range(n)]for i in range(n)]
    for i in range(n):
        for j in range(n):
            matT[i][j]=mat[j][i]
            
    printMatriz(matT)
    
##############################################################

main()
```

```python
#025.: Escreva um programa que leia todas as posições de uma matriz nxn e que mostre as posições não nulas da matriz.

def leMatriz():
    n=int(input("Digite a dimensão da matriz quadrada: "))
    mat=[[]for i in range(n)]
    
    for i in range(n):
        for j in range(n):
            aux=float(input('Mat[{}][{}]='.format(i+1, j+1)))
            mat[i].append(aux)
    
    return mat

mat=leMatriz()
n=len(mat)
acu=0
print()
for i in range(n):
    for j in range(n):
        if mat[i][j]!=0:
            print("[{}][{}] é uma posição não nula".format(i+1, j+1))
            acu=acu+1
            
print()
print("A matriz possui {} posições não nulas".format(acu))
```
